```python
def getratings(title):
    res = movie.search(title)
    if len(res) == 0:
        return np.NaN
    return res[0].vote_average

data['vote_average'] = data['title'].swifter.apply(getratings)
```

```python
df = pd.read_csv('/kaggle/input/withvote/withvote.csv')
df
```

```python
high = df[df['vote_average'] > 6.5]
high
```

```python
high.reset_index(drop = True, inplace = True)
```

```python
high
```

```python
high.to_csv('highrated1.csv')
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(high['final'])
similar = cosine_similarity(count_matrix)

def recommend(movie):
    movie_idx = high[high['title'] == movie].index[0]
    distances = similar[movie_idx]
    #to get indices of movies 
    distances_with_idx = enumerate(distances)
    #sorting based on similarities which is 1st index of tuple
    similar_list = sorted(list(distances_with_idx), reverse = True, key = lambda x : x[1])
    top_list = similar_list[1:11] 
    
    for i in top_list:
        print(high.loc[i[0]].title)

recommend('Avengers: Endgame')
```

```python
pickle.dump(similar, open('similarity.pkl', 'wb'))
```
