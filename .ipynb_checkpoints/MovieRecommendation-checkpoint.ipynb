```python
start = time()
movies['crew'] = movies['crew'].map(lambda x : ast.literal_eval(x))
movies['cast'] = movies['cast'].map(lambda x : ast.literal_eval(x))
movies['genres'] = movies['genres'].map(lambda x : ast.literal_eval(x))
end = time()
print(end - start)
print(type(movies.loc[0]['crew']))
print(type(movies.loc[0]['cast']))
print(type(movies.loc[0]['genres']))
```

```python
movies.loc[0]['cast']
```

```python
# we gotta grab the top 3 actors

def actor(ls, pos):
    cast = []
    for item in ls:
        cast.append(item['name'])
    if len(cast) >= pos:
        return cast[pos - 1]
    else:
        return np.NaN
```

```python
movies['actor1'] = movies['cast'].map(lambda x : (actor(x, 1)))
movies['actor2'] = movies['cast'].map(lambda x : (actor(x, 2)))
movies['actor3'] = movies['cast'].map(lambda x : (actor(x, 3)))
movies.head(10)
```

```python
movies.loc[0]['crew']
```

```python
#From the crew series we can observe that the first ID is always of the Director. Let's extract that!
def director(ls):
    if len(ls) == 0:
        return np.NaN
    dic = ls[0]
    name = dic['name']
    if 'name' in dic and name != '':
        return name
    else:
        return np.NaN

movies['director'] = movies['crew'].map(lambda x : director(x))
movies.head(10)
```

```python
def genres_list(ls):
    gen = []
    s = " "
    for item in ls:
        if 'name' in item and item['name'] != '':
            if item['name'] == 'Science Fiction':
                gen.append('Sci-Fi')
            else:
                gen.append(item['name'])
        else:
            continue
    if len(gen) == 0:
        return np.NaN
    else:
        s = ' '.join([str(elem) for elem in gen])
        return s
```

```python
movies['genres'] = movies['genres'].map(lambda x : genres_list(x))
movies.head(10)
```

```python
movies.isnull().sum()
```

```python
#removing missing values
movies.dropna(subset = ['overview', 'director', 'genres', 'actor1'], inplace = True)
movies.reset_index(drop = True, inplace = True)
movies.isnull().sum()
```

# Overview Processing


```python
nltk.download('popular')
stop = stopwords.words("english")


lem = WordNetLemmatizer()
def convert_emojis(text):
    for emot in UNICODE_EMOJI:
        text = text.replace(emot, "_".join(UNICODE_EMOJI[emot].replace(",","").replace(":","").split()))
    return text


 


from nltk.corpus import wordnet
 
lemmatizer = WordNetLemmatizer()
 

def pos_tagger(nltk_tag):
    if nltk_tag.startswith('J'):
        return wordnet.ADJ
    elif nltk_tag.startswith('V'):
        return wordnet.VERB
    elif nltk_tag.startswith('N'):
        return 'n'
    elif nltk_tag.startswith('R'):
        return wordnet.ADV
    else:         
        return None


def clean_text(s):
    lowered_text = s.lower()
    # Handling html text
    sentence = cleanhtml(lowered_text)
    # Handling underscore and -
    sentence = re.sub("-|_", " ", sentence)
    
    # Extending abbrevations like he's
    sentence  = contractions.fix(sentence)
    # Handling newline and tab space charecters
    sentence = re.sub('\n|\t', "", sentence)
    """
    For Typos 
    But slows the processing many times
    sentence = str(TextBlob(sentence).correct())
    """
    # Handling symbols (mainly for puncs)
    no_punc = clean_punc(sentence)
    # Getting the pos_tags after tozenizing
    l = nltk.pos_tag(nltk.word_tokenize(no_punc))
    # Converting pos_tag to approriate format for lemmatizer to work with
    l = [(word,pos_tagger(tag)) for (word, tag) in l if (word not in stop) and (not word.isdigit())]
    # Using lemmatizer according to the pos_tags for proper lemmatizing
    useful = " ".join([lem.lemmatize(i, pos=j) for i, j in l if j != None])
    # Converting emoji to text
    useful = convert_emojis(useful)
    return useful.strip()
def cleanhtml(sentence):
  exp = re.compile("<.*?>")
  cleantext = exp.sub(" ", sentence)
  return cleantext
 
def clean_punc(text):
  exp =('[%s]' % re.escape(module_string.punctuation))
  return re.compile(exp).sub('', text)
text = "<div> </div>This is a text!!! It has (parenthesis), square and  curly brackets [[{{}}]]  and hashtags #.-_$ 20 😍"
clean_text(text)
```

```python
start = time()
movies['processed_overview'] = movies['overview'].swifter.apply(clean_text)
end = time()
print(end - start)
```
